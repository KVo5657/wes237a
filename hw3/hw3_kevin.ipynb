```python
# import ctypes interface for Python
import ctypes
import os
import time
```

```python
cycletime = ctypes.CDLL('./cycletime.so')
```

```python
# Access the shared library functions using the ctypes module, 
# but donâ€™t wrap the function calls in a python function.

# One function to intialize the PMU counters
cycletime.init_counters(1, 0);

# One function to get the cycle count
cycletime.get_cyclecount()
```

# Homework Part 2: Comparing and Gathering Data
In this section, we are going to use psutil to monitor CPU usage in percent, and the time module and PMU counting to evaluate the recursive fibonacci sequence timing operations.

```python
## get before time using the python time module
# start = time.time()
# print("Time Before:", start)

# CPU Number input
cpu_str = input("Enter your cpu number: ")
cpu = int(cpu_str)

# CPU
print(f"CPU you use is CPU {cpu}.")
n = input("Enter n for fib.py: ").strip()

# Initialize the cyclecounter
cycletime.init_counters(1, 0);

# get the before cycle count
cpu_count_before = cycletime.get_cyclecount();
# print("Cycle Count Before:", cpu_count_before)

# Run the File Using Inputs
os.system(f"printf '{n}\n' | taskset -c {cpu} python3 fib_hw3.py") 

## get after time using the python time module
# end = time.time()
# print("Time After:", end)

# get the after cycle count
cpu_count_after = cycletime.get_cyclecount();
# print("Cycle Count After:", cpu_count_after)

# cycle count
cpu_count = cpu_count_after - cpu_count_before;
# print("Cycle Count:", cpu_count)

## time elapsed
# length = end-start

## Show the results : this can be altered however you like
# print("It took", length, "seconds!")

# Ordered Print For Easy Transfer to Spreadsheet
print("- Cycle Count Before:", cpu_count_before, "cycles")
print("- Cycle Count After:", cpu_count_after, "cycles")
print("- Cycle Count:", cpu_count, "cycles")
```
